# Fantasy Premier League predictions project

## Introduction

Fantasy Premier League is a fantasy sports game played by five million people worldwide. Game-players choose a team of English Premier League footballers, who receive points based on their performances each week. The aim is to maximise points scored by your selection under a budgetary constraint, with the final aim of earning more points than other players.

## Project Objective

The over-arching analysis question on this data-set is: To what extent can past weeks’ data help decision-making during the year’s competition?
Relevant analysis areas are: building a predictive model for footballers’ points scores in upcoming weeks, and identifying winning strategies while taking into account the budgetary and positional constraints.
The primary objective will be to build a predictive model, with a secondary objective of exploring the data and identifying game strategies.

## Data source

The data-set is due to https://github.com/JoGall/FantasyPL, scraped from the official site in 2016/17. It contains 54 columns and 23,679 rows, where each row is a player’s data for one gameweek. In total there is data for 683 footballers over 38 gameweeks.

The data is sourced from the official website, so little data cleaning is necessary.

## Analysis Plan

1. Assess whether a predictive model for points is feasible.
2. Build predictive model, looking at number of weeks to forecast, and how soon in the season a model can begin to be used.
3. Explore strategies for success in the game: when to pick players with regards to their fixtures, identify differences in points awarded by player type and position; since the game is played under a budgetary constraint (gameplayers have £100m to spend on players), consider how best to spread the allocated budget amongst positions, high-value players, etc.

## Data Pre-processing

The csv data is read into a pandas dataframe, removing the empty “ea_index” column, and observations relating to players who played 0 minutes during the season. These players artificially increased the prediction scores, as a prediction of 0 points for players playing 0 minutes is trivial.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set() # set seaborn style for charts

# Import data
df = pd.read_csv('16-17_formatted.csv')
df.drop('ea_index', axis=1, inplace=True)

# Remove observations with very few points/events (0 minutes in total)
df_mins = df.pivot_table(index=['player_id'],values=['minutes'],aggfunc=np.sum)
idx_mins = df_mins!=0
id_nonZeroMins = np.where(idx_mins)[0] +1 # Add 1 to account for zero-indexing of arrays
df = df.loc[df['player_id'].isin(id_nonZeroMins)]

# Identify which variables are events and which are statistics (i.e. which can be summed over multiple gameweeks)
varnames = list(df.columns.values)
eventVars = ['team_goals','opposition_goals','total_points','minutes','goals_scored','assists','clean_sheets','goals_conceded','own_goals','penalties_saved','penalties_missed','yellow_cards','red_cards','saves','bonus','bps','open_play_crosses','big_chances_created','clearances_blocks_interceptions','recoveries','key_passes','tackles','winning_goals','attempted_passes','completed_passes','penalties_conceded','big_chances_missed','errors_leading_to_goal','errors_leading_to_goal_attempt','tacked','offside','target_missed','fouls','dribbles','influence','creativity','threat','ict_index','ea_index']
statVars = ['value','transfers_balance','selected','transfers_in','transfer_out']
```

Separately, in Excel, I created a ‘fixture difficulty’ variable, to account for the fact the footballers in the top teams will score more points against weaker teams than against other top teams.
I intended to use a simple class variable indicating whether the fixture was against a ‘top six’ team (Arsenal, Chelsea, Liverpool, Manchester City, Manchester United, Tottenham). However, as the model looks at difficulty of past fixtures to predict future points, we require a variable for which we can calculate the average of previous fixture difficulties.

## Building the Predictive Model

As I want to look at both the number of predictable gameweeks and the earliest time a predictive model is useful, I split the data into multiple sets:

$data_{ij}$ = { data: beginning at gameweek i which predicts j gameweeks ahead }

```python
from collections import defaultdict

events_to_X = {}
averaged_vars = {}
variables = {}
pts = defaultdict(dict)

for gw in range(1,38):
    # Create dictionary of events up to GW X
    events_to_X[gw] = pd.pivot_table(df.loc[df['round']<=gw],index=['name','player_id','pos'],values=eventVars,aggfunc=np.sum)
    averaged_vars[gw] = pd.pivot_table(df.loc[df['round']<=gw],index=['name','player_id','pos'],values=['fixture_difficulty','value'])
    variables[gw] = pd.merge(events_to_X[gw],averaged_vars[gw],left_index=True,right_index=True)
    
    # Create dictionary of points scores, starting from 'GW', looking ahead 'weeks' number of GWs.
    for weeks in range(1,39-gw):
        try:
            pts[gw][weeks] = pd.pivot_table(df[(df['round']>gw)&(df['round']<=gw+weeks)],index=['name','player_id','pos'],values=['total_points'],aggfunc=np.sum)
        except:
            continue
```

### Linear Regression
After separating training and test data for each data-set, I fit a linear regression model to the training data, and evaluated the R2 score for the models: in total there are 703 models trained, one for each combination of ‘starting gameweek’ and ‘weeks ahead’.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scores_train = defaultdict(dict)
scores_test = defaultdict(dict)

# Separate data into training and test data, and standardise features; one set for each combination of GW and 'weeks ahead'

for gw in range(1,38):
    for weeks in range(1,39-gw):
        
        # Merge dependent and independent variables
        try:
            data = pd.merge(variables[gw],pts[gw][weeks],left_index=True,right_index=True)
        except:
            continue
        
        # Split into training and test data
        X = data.iloc[:,:-1]
        y = data.iloc[:,-1]
        
        # Standarize features
        scaler = StandardScaler()
        X_std = scaler.fit_transform(X)
        
        # Separate data
        X_train, X_test, y_train, y_test = train_test_split(X_std,y, test_size=0.25,random_state=42)
        cols = X_train.shape[1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, median_absolute_error

# Fit model and predict
for gw in range(1,38):
    for weeks in range(1,39-gw):
        mdl = LinearRegression(fit_intercept=True)
        
        mdl.fit(X_train,y_train)
        train_prediction = mdl.predict(X_train)
        test_prediction = mdl.predict(X_test)
        
        # Evaluate model performance
        scores_train[gw][weeks] = r2_score(y_train,train_prediction)
        scores_test[gw][weeks] = r2_score(y_test,test_prediction)
```

```python
# Plot scores on training data
scores_train_df = pd.DataFrame.from_dict(scores_train,orient='index')        
scores_train_by_gw = scores_train_df.mean(axis=0)
scores_train_by_weeksAhead = scores_train_df.mean(axis=1)

plt.plot(scores_train_by_weeksAhead)
plt.plot(scores_train_by_gw)
plt.legend(('# of weeks ahead','starting GW'),loc='upper right')
plt.ylabel('R2')
plt.title('Linear Regresssion - train')
plt.show()

# Plot scores on test data
scores_test_df = pd.DataFrame.from_dict(scores_test,orient='index')        
scores_test_by_gw = scores_test_df.mean(axis=0)
scores_test_by_weeksAhead = scores_test_df.mean(axis=1)

plt.plot(scores_test_by_weeksAhead)
plt.plot(scores_test_by_gw)
plt.legend(('# of weeks ahead','starting GW'),loc='upper right')
plt.ylabel('R2')
plt.title('Linear Regression - test')
plt.show()
```

The above charts show the performance of the simple linear regression model. 

One can see that a consistent R2 score of 0.4 is achieved on the training data, but this reduces to around 0.2 on the test data, indicating the models are overfitting. 

It is also noted that the general trend is for performance to improve as the season progresses, and the greater the number of weeks I am predicting. The “sweet spot” appears to be around 5-15 weeks ahead, and starting at week 5-10.

As there are a large number of features in the dataset, I applied Lasso regularisation to reduce the effect of ‘low importance’ features, using grid search to find the optimal hyper-parameter, $\alpha$. The best-performing model had $\alpha$=0.01.

```python
# Fit model and predict
from sklearn.linear_model import Lasso

for gw in range(1,38):
    for weeks in range(1,39-gw):
        mdl = Lasso(alpha=0.01)
        
        mdl.fit(X_train,y_train)
        train_prediction = mdl.predict(X_train)
        test_prediction = mdl.predict(X_test)
        
        # Evaluate model performance, and feature importance
        scores_train[gw][weeks] = median_absolute_error(y_train,train_prediction)
        scores_test[gw][weeks] = median_absolute_error(y_test,test_prediction)
```

```python
# Plot scores on training data
scores_train_df = pd.DataFrame.from_dict(scores_train,orient='index')        
scores_train_by_gw = scores_train_df.mean(axis=0)
scores_train_by_weeksAhead = scores_train_df.mean(axis=1)

plt.plot(scores_train_by_weeksAhead)
plt.plot(scores_train_by_gw)
plt.legend(('# of weeks ahead','starting GW'),loc='upper right')
plt.ylabel('R2')
plt.title('Lasso - train')
plt.show()

# Plot scores on test data
scores_test_df = pd.DataFrame.from_dict(scores_test,orient='index')        
scores_test_by_gw = scores_test_df.mean(axis=0)
scores_test_by_weeksAhead = scores_test_df.mean(axis=1)

plt.plot(scores_test_by_weeksAhead)
plt.plot(scores_test_by_gw)
plt.legend(('# of weeks ahead','starting GW'),loc='upper right')
plt.ylabel('R2')
plt.title('Lasso - test')
plt.show()
```

### Random Forest

I next looked at a model which is less affected by correlations between features: A Random Forest model. In addition, Random Forest models can model a non-linear relationship between variables.

```python
# Fit model and predict
from sklearn.ensemble import RandomForestRegressor

for gw in range(1,38):
    for weeks in range(1,39-gw):
        mdl = RandomForestRegressor(oob_score=False)
        
        mdl.fit(X_train,y_train)
        train_prediction = mdl.predict(X_train)
        test_prediction = mdl.predict(X_test)
        
        # Evaluate model performance, and feature importance
        scores_train[gw][weeks] = median_absolute_error(y_train,train_prediction)
        scores_test[gw][weeks] = median_absolute_error(y_test,test_prediction)
```

```python
# Plot scores on training data
scores_train_df = pd.DataFrame.from_dict(scores_train,orient='index')        
scores_train_by_gw = scores_train_df.mean(axis=0)
scores_train_by_weeksAhead = scores_train_df.mean(axis=1)

plt.plot(scores_train_by_weeksAhead)
plt.plot(scores_train_by_gw)
plt.legend(('# of weeks ahead','starting GW'),loc='upper right')
plt.ylabel('R2')
plt.title('Random Forest - train')
plt.show()

# Plot scores on test data
scores_test_df = pd.DataFrame.from_dict(scores_test,orient='index')        
scores_test_by_gw = scores_test_df.mean(axis=0)
scores_test_by_weeksAhead = scores_test_df.mean(axis=1)

plt.plot(scores_test_by_weeksAhead)
plt.plot(scores_test_by_gw)
plt.legend(('# of weeks ahead','starting GW'),loc='upper right')
plt.ylabel('R2')
plt.title('Random Forest - test')
plt.show()
```

```python
from sklearn.feature_selection import SelectFromModel

scores_train_important = defaultdict(dict)
scores_test_important = defaultdict(dict)
features = defaultdict(dict)

for gw in range(1,38):
    for weeks in range(1,39-gw):
        # Train model on important features only
        sfm = SelectFromModel(mdl, threshold='median')
        sfm.fit(X_train, y_train)
        
        X_train_important = sfm.transform(X_train)
        X_test_important = sfm.transform(X_test)
        
        mdl_important = RandomForestRegressor(oob_score=False)
        mdl_important.fit(X_train_important, y_train)
        
        train_prediction_important = mdl_important.predict(X_train_important)
        test_prediction_important = mdl_important.predict(X_test_important)
        
        scores_train_important[gw][weeks] = r2_score(y_train,train_prediction_important)
        scores_test_important[gw][weeks] = r2_score(y_test,test_prediction_important)
        features_important = mdl.feature_importances_
        features[gw][weeks] = list(zip(data.columns,features_important))
```

I investigated further whether some features are better estimates than others, selecting based on their ‘importance score’, which in regression problems is the variance of the feature. After selecting all features with higher-than-median importance score, the results were as follows:

There is very little difference between Random Forest models trained on all features vs. those trained on a subsection of features. 

The feature importance scores from these models are shown below. Predictably, ‘total points’ in previous weeks is the top feature, and perhaps surprisingly ‘goals scored’ is one of the least important - likely due to it being a generative event for ‘total points’.

```python
def flatten(features):
    """
    Flatten feature importance scores into dataframe, return sorted list of features by average importance across all models
    
    Input: nested dictionary of feature scores
    Output: flattened dataframe of feature scores
    """
    j = 0
    f_score = {}
    f_array = np.zeros((703,37))
    f_rowName = []
    for gw in range(1,38):
        for weeks in range(1,39-gw):
            for i in range(37):
                key = (gw, weeks)
                f_score[key] = features[gw][weeks]
                scores = [i[1] for i in f_score[key]]
                scores_array = np.asarray(scores).reshape(1,37)
                f_array[j] = scores_array
            f_rowName.append(key)
            j+=1
    columnNames = [i[0] for i in features[1][1]]
    f_df = pd.DataFrame(f_array,columns=columnNames,index=f_rowName)
    return f_df

# Convert nested dictionary to list
f_df = flatten(features)

# Plot average Gini score on a bar chart
f_mean = np.mean(f_df)
f_mean.sort_values(ascending=False, inplace=True)
f_mean.plot.bar()
plt.show()

importances = mdl.feature_importances_
indices = np.argsort(importances)[::-1]
names = [data.feature_names[i] for i in indices]
names = list(data.columns.values[indices])
plt.bar(range(cols),importances[indices])
plt.xticks(range(cols), names, rotation=90)
plt.show()
```

## Assessing the Model in Practice

Building a predictive model does not in itself guarantee success. A player must maximise points by transferring footballers into the team at the correct moment, taking advantage of form, favourable fixtures, and value.
In this section I will test the model by using it to choose footballers and transfers throughout a season. As game strategy will affect a player’s success, and there are many valid options, I will use a simple strategy, outlined below.

Strategy:
1. Identify a gameweek to begin the experiment.
2. List all footballers by their predicted points in upcoming gameweeks.
3. Choose players for the team by value and position under the game’s constraints using a “first-fit” bin-packing algorithm.
4. Compare predicted points to points achieved.
5. Allow 1 transfer per gameweek by assessing predicted points in upcoming gameweeks.

### Step 3

This step is where much more work would need to be done: we have a complex optimisation problem. As a demonstration of how the ‘first-fit’ algorithm fails I will show a few steps of the algorithm.

Note that 15 footballers must be chosen: 2 Goalkeepers, 5 Defenders, 5 Midfielders, and 3 Forwards. Each week 11 from 15 are chosen by the player to be in their ‘first team’, whose points will count towards the player’s total, with the remaining considered to be ‘on the bench’. The budget for all 15 footballers is 100.

The first ten footballers above have a combined value of 85. The remaining five footballers must have a value of 15. The minimum value for any player in the game is 4. Therefore, we require a decision-rule to retract a choice so that the full 15 footballers can be chosen.

I suggest placing limits on the value by position. For example, a combined value of 35 for midfielders, 30 for forwards, 10 for goalkeepers, and 25 for defenders. However, assessing this strategy would require a full exploration of winning strategies in the game, requiring data from a large proportion of the 5 million game-players.
An alternative would be to investigate different optimisation algorithms.

### Further Research
#### Model Improvements
The first fruitful research area is improving the model and testing with further data. For this project I used data from the 2016/17 football season. If data from previous seasons (or the current) could be used, one could ensure that any built model doesn’t depend on performance outliers. An example of a performance outlier is the 5000/1 outsider, Leicester City, winning the Premier League in 2015/16.
The model might be improved with additional features, such as the new “expected goals (xG)” statistic, or with a ‘points per £cost’ derived feature.
#### Strategy
Aside from model improvements, an interesting research area would be the strategic side of the game: developing decision rules for picking a team. On this project we used a first-fit algorithm to demonstrate that much more work is needed in this area. If one could access data from the 5 million players of Fantasy Premier League each week, one could investigate the formations, number of transfer, and other strategies which increase points totals.
#### Form
Part of the challenge of the game is identifying footballers when they are “on form” i.e. they play well consistently, rather than the occasional successful performance. Identifying these moments early would allow a player can bring them in to their team for their ‘good patches’ and remove them when the ‘good patch’ ends. For this project we concentrated on a simple prediction of points. It would be interesting to see whether one could predict a ‘run of form’ early on, or at least identify the key features of a player being ‘in form’.
